```python
# importing required libraries
import seaborn as sns
#sns.set()
sns.set(style="darkgrid")


import numpy as np
import pandas as pd

# importing matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
plt.rcParams['figure.figsize']=(10,10)
```

```python
#Use the “Titanic” dataset from Seaborn and do the following:
data = sns.load_dataset("titanic")
# Drop rows with missing values for simplicity in some plots
data = data.dropna(subset=['age', 'fare', 'embarked'])

data.head()
```

```python
#1. Countplot: Passenger Count by Embarked Port
#● Create a countplot to display the number of passengers who boarded from each
plt.figure(figsize=(8,5))
sns.countplot(data, x='embarked', hue='sex')
plt.title('Passenger Count by Embarked Port (Separated by Gender)')
plt.xlabel('Embarked Port')
plt.ylabel('Count')
plt.legend(title='Sex', fontsize=8)
plt.show()
```

Insights:

*  Most passengers boarded from port S (Southampton).
*   There are more male passengers than female across all ports.
*  Port C (Cherbourg) shows a relatively higher proportion of female passengers compared to others.









```python
#2. Swarmplot: Age Distribution across Classes
#● Create a swarmplot to visualize the distribution of age across passenger classes (pclass), categorized by sex.
plt.figure(figsize=(8,5))
sns.swarmplot(data, x='pclass',y='age', hue='sex')
plt.title('distribution of age across passenger classes (Separated by Gender)')
plt.xlabel('pclass')
plt.ylabel('age')
plt.legend(title='Sex', fontsize=8)
plt.show()
```

Insights:


*   First-class passengers tend to be older on average.
*   Third-class shows a wider spread, including many younger passengers.
*   Gender distribution appears fairly mixed across ages.






```python
#3. Pie Chart: Gender Distribution
#● Create a pie chart using Matplotlib to show the percentage of male vs. female passengers.- Add percentage labels using autopct- Use explode for one slice to
#highlight it.
gender_counts = data['sex'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%',
        explode=[0.1, 0], colors=['lightblue', 'lightcoral'], startangle=90)
plt.title('Gender Distribution of Passengers')
plt.axis('equal')
plt.show()

```

Insights:


*   Male passengers make up the majority (~64%).
*   Female passengers represent a smaller but significant portion.
*   The dataset is gender-imbalanced toward males.






```python
#4. Heatmap: Correlation Matrix
#● Create a heatmap to show correlations between age, fare, pclass, sibsp, parch,and survived. (Use seaborn.heatmap with annotations)
plt.figure(figsize=(8,6))
corr = data[['age', 'fare', 'pclass', 'sibsp', 'parch', 'survived']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

Insights:


*   Fare and pclass are strongly negatively correlated (higher class → higher fare).
*   Survival is negatively correlated with pclass (higher class → better survival).
*   Weak correlations exist between age and survival.













```python
#5. Violin Plot: Fare by Class and Gender
#● Create a violin plot to visualize fare distribution across classes, separated by gender.
plt.figure(figsize=(10,6))
sns.violinplot(data, x='pclass', y='fare', hue='sex', split=True)
plt.title('Fare Distribution by Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

Insights:


*  First-class fares are significantly higher and more spread out.
*  Third-class fares are low and tightly clustered.
*   Slight variation between male and female fare distributions within classes.






```python
#6. Scatter Plot: Age vs Fare
#● Create a scatter plot to explore the relationship between age and fare, colored by survival status.
#● Use a seaborn scatterplot with hue='survived'. Each plot must have titles, labels,and legends where needed.
plt.figure(figsize=(10,6))
sns.scatterplot(data, x='age', y='fare', hue='survived')
plt.title('Age vs Fare (Colored by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived')
plt.show()
```

Insights:


*   Passengers who paid higher fares had a higher chance of survival.
*   Many low-fare passengers did not survive.
*   Age alone does not show a strong pattern with survival.





###Summary Insight


* Class and fare were major factors influencing survival.
* Gender imbalance exists, but survival trends often favored women.
* Socioeconomic status (class/fare) had a stronger impact than age.
